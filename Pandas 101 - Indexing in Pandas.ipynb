```python
df = pd.read_csv("world_population.csv", index_col = "Country")

df
```

```python
df.reset_index(inplace=True)
```

```python
df
```

- `df.set_index('Country', inplace=True)`

- `set_index('Country')`

Takes the column named 'Country' and makes it the row index (the labels on the left).
That column is removed from the columns and used as the index instead.
Other columns stay as columns; only which values identify each row changes.

- `inplace=True`

Tells pandas to change df itself instead of returning a new DataFrame.
After the call, df is updated; you don’t assign the result to a variable.
With inplace=False (the default), you’d get a new DataFrame and the original df would be unchanged unless you did df = df.set_index('Country').
Example

Before: index might be 0, 1, 2, ... and there’s a Country column.

- After `df.set_index('Country', inplace=True)`:

Rows are labeled by country name (e.g. "Albania", "France").
There is no longer a separate Country column.
That’s why operations like df.filter(items=["Albania"], axis=0) work when "Albania" is an index label after you’ve set the index to country.

# Why filter gives nothing
- `filter(items=["Africa", "Angola"], axis=0)` looks for row index labels that equal "Africa" or "Angola".
With a MultiIndex, each row’s label is usually a tuple like `("Africa", "Angola")`. 
- That tuple is not the same as the string "Africa" or "Angola", so nothing matches and you get an empty frame.



```python
df.filter(items = ["Africa","Angola"], axis =0)
```

```python
df.iloc[1]
```
