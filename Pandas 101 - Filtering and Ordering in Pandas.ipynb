# Filtering and Ordering

```python
import pandas as pd
```

```python
df = pd.read_csv("world_population.csv")
df.head()
```

```python
df.loc[:,["Country","Continent"]]
```

```python
df.iloc[:,1:3].head()
```

```python
df.loc[df["Continent"] == "Asia",df.columns.str.contains("C")].head()
```

```python
df.loc[:, df.columns.str.contains("CC")]
```

```python
df.loc[ (df['Rank'] <= 10) & (df['Capital'] == "Dhaka")]
```

```python
specific_countries = ['Bangladesh','Brazil']
temp = df.iloc[df['Country'].isin(specific_countries), 2:5]
temp["my_rank"] = 0
temp["my_rank"] = 2*df["Rank"]
temp
```

```python
df[~df['Country'].isin(specific_countries)]
```

```python
#this is the opposite
df[~df['Country'].isin(specific_countries)].head()
```

```python
df[df['Country'].str.contains('United')]
```

```python
df['Country'].is_unique
```

```python
df.loc[df['Country'].str.contains("Algeria"),:]
```

```python
my_countries = ["Afghanistan","Yemen"]
df.loc[df['Country'].isin(my_countries),:]
```

```python
df.head()
```

```python
#after setting Country as index, we can use like, items with list to fiter rows
df2 = df.set_index('Country')
df2
```

```python
#"Algeria" is one of the value in index
df2.loc["Algeria"]
```

```python
df2.filter(items = ['Zimbabwe'], axis=0)
```

```python
df2.filter(like = 'United', axis = 0)
```

```python
df2.filter(items = ['Continent','CCA3'], axis = 1)
```

```python
df2.iloc[:, df2.columns.str.contains("Population")]
```
